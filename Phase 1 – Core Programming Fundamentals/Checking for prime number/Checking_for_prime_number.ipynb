```python
import math
```

```python
n = int(input("enter a number:"))
print(n)
```

```python
if n <= 1:
    print(f"{n} is not a prime number")
else:
    is_prime = True
    for i in range(2, int(math.sqrt(n)) + 1):
        if n % i == 0:
            is_prime = False
            break
    if is_prime:
        print(f"{n} is a prime number")
    else:
        print(f"{n} is not a prime number")
```
